```python
import pandas as pd
import glob
#import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
data= pd.read_csv("../input/9-subjects-train-data/Train_9subjects.csv")
data.shape
```

```python
X,y = data.loc[:,data.columns!="Choice"],data.Choice
X.shape,y.shape
```

```python
X.head(),y.head()
```

```python
from sklearn.utils import shuffle
X_sh,y_sh = shuffle(X,y,random_state=9896)


from sklearn.preprocessing import LabelEncoder
e = LabelEncoder()
y_enc = e.fit_transform(y_sh)

#y_enc[0:10]

th_data = X_sh[["RAW_AF7","RAW_AF8","RAW_TP9","RAW_TP10"]]

X_train = th_data.values.reshape(277052,4,1)                                           # Reshaping the data for CNN
import numpy as np

X_train = np.asarray(X_train)

x_sh = th_data.fillna(0)
#X_train.shape


from keras import models,layers

model = models.Sequential()
model.add(layers.BatchNormalization(input_shape=(4,)))
model.add(layers.Dense(128,activation='relu',kernel_regularizer='l1'))
model.add(layers.BatchNormalization())
model.add(layers.Dense(512,activation='relu',kernel_regularizer='l1'))

model.add(layers.Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['acc'])

model.fit(x_sh,y_enc,epochs=5)
```

```python
from sklearn.utils import shuffle
X_sh,y_sh = shuffle(X,y,random_state=9896)


from sklearn.preprocessing import LabelEncoder
e = LabelEncoder()
y_enc = e.fit_transform(y_sh)

#y_enc[0:10]

th_data = X_sh[["Theta_AF7","Theta_AF8","Theta_TP9","Theta_TP10"]]

X_train = th_data.values.reshape(277052,4,1)                                           # Reshaping the data for CNN
import numpy as np

X_train = np.asarray(X_train)

x_sh = th_data.fillna(0)
#X_train.shape


from keras import models,layers

model = models.Sequential()
model.add(layers.BatchNormalization(input_shape=(4,)))
model.add(layers.Dense(128,activation='relu',kernel_regularizer='l2'))
model.add(layers.BatchNormalization())
model.add(layers.Dense(512,activation='relu',kernel_regularizer='l2'))

model.add(layers.Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['acc'])

model.fit(x_sh,y_enc,epochs=10)
```
